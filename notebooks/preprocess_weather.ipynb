```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
weather = pd.read_pickle('../input/weather_only.pkl')
```

```python
weather
```

```python
weather = weather.reset_index(inplace=False)
weather = weather[weather.index % 2 == 0].reset_index(drop=True, inplace=False)
weather = weather[::-1].reset_index(drop=True, inplace=False)
weather
```

```python
weather = weather.rename(columns={'Local time in Tallinn ': 'Timestamp'})
weather['Timestamp'] = pd.to_datetime(weather['Timestamp'], format='%d.%m.%Y %H:%M')
weather['Timestamp'] = weather['Timestamp'].dt.round('h')
weather
```

```python
# 1. Make a safe copy, leave NaNs as they are
ww_series = weather['WW']

# 2. Process only non-null rows
non_null_ww = ww_series.dropna().apply(lambda x: [item.strip() for item in x.split(',')])

# 3. Fit and transform only non-null data
mlb = MultiLabelBinarizer()
mlb_encoded = pd.DataFrame(mlb.fit_transform(non_null_ww), 
                           index=non_null_ww.index, 
                           columns=mlb.classes_)

# 4. Create a full DataFrame with all rows and fill missing ones with zeros
weather_encoded = pd.DataFrame(0, index=weather.index, columns=mlb.classes_)
weather_encoded.update(mlb_encoded)

# 5. Concatenate with original or drop original column
weather = pd.concat([weather, weather_encoded], axis=1)
weather
```

```python
weather = weather.drop(columns=['DD', 'ff10', 'WW', "W'W'", 'c', 'VV'])
weather.head()
```

```python
weather = weather.set_index('Timestamp', inplace=False)

expected_range = pd.date_range(start='2023-01-01 00:00', end='2023-12-31 23:00', freq='h')
actual_range = weather.index

missing = expected_range.difference(actual_range)
print(missing)

duplicates = weather.index[weather.index.duplicated()]
print(duplicates)
```

```python
weather = weather[weather.index < '2024-01-01 00:00']
weather.tail()
```

```python
# Handle missing timestamps by adding them to the dataframe with NaN values
for timestamp in missing:
    weather.loc[timestamp] = np.nan

# Handle duplicated timestamps - keep the first occurrence
weather = weather[~weather.index.duplicated(keep='first')]

# Sort the index to ensure timestamps are in order
weather = weather.sort_index()

# Display the resulting dataframe
print(f"Shape after handling missing and duplicated timestamps: {weather.shape}")
weather
```

```python
weather.interpolate(method='linear', inplace=True, limit_direction='both')
weather = weather.reset_index(inplace=False)
weather
```

```python
weather['temp_diff_1h'] = weather['T'] - weather['T'].shift(1)
weather['humid_diff_1h'] = weather['U'] - weather['U'].shift(1)
weather.fillna(0, inplace=True)
weather
```
