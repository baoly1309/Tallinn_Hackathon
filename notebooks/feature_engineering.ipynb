```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_pickle('../input/data.pkl')
```

```python
df
```

```python
df['month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)
df['day_sin'] = np.sin(2 * np.pi * df['Day_of_month'] / 31)
df['day_cos'] = np.cos(2 * np.pi * df['Day_of_month'] / 31)
df['weekday_sin'] = np.sin(2 * np.pi * df['Day_of_week'] / 7)
df['weekday_cos'] = np.cos(2 * np.pi * df['Day_of_week'] / 7)
df['hour_sin'] = np.sin(2 * np.pi * df['Hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['Hour'] / 24)
```

```python
df_new = df.drop(columns=['Month', 'Day_of_month', 'Day_of_week', 'Hour'])
df_new
```

```python
df_new.to_pickle('../input/data_clean.pkl')
```
